```python
from bs4 import BeautifulSoup
from selenium import webdriver

driver= webdriver.Chrome('/Users/hyewon_oh/study/chromedriver')
driver.get('https://www.aladin.co.kr/shop/common/wbest.aspx?BranchType=1')

rank = 0
j=3

date_list=[]
    
for i in range(20):
    source = driver.page_source

    s1 = BeautifulSoup(source, "html.parser")
    
    title_list=s1.select(' div.ss_book_list > ul > li > a >  b')
    author=s1.select("a[href^='/Search/wSearchResult.aspx?AuthorSearch=']")
    publish=s1.select("a[href^='/search/wsearchresult.aspx?PublisherSearch']")
    price=s1.select('span.ss_p2 > b > span')
    date=s1.select('div.ss_book_list > ul > li ')
    
    book = s1.find_all('div', class_='ss_book_box')
    
    for item in book:

        if item.find('span', class_='ss_ht1'):
            date = item.find_all('li')[2].text.split('|')[2].strip()
        else:
            date = item.find_all('li')[1].text.split('|')[2].strip()
        date_list.append(date)
        
    for idx in range(len(title_list)):
        rank +=1
        print('%s위 도서정보' %rank)
        print(title_list[idx].text.strip())
        print(author[idx].text.strip())
        print(publish[idx].text.strip())
        print(price[idx].text.strip())
        print(date_list[idx])
        print("-------------------------------")
        
    n=j+i
    if i == 19 :
        driver.close()
    elif i == 9 :
        driver.find_element_by_xpath('//*[@id="navi_next"]').click()
        driver.find_element_by_xpath('//*[@id="newbg_body"]/div[3]/ul/li[%s]/a'%n).click()
    else :
        driver.find_element_by_xpath('//*[@id="newbg_body"]/div[3]/ul/li[%s]/a'%n).click()
```
