```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time

driver= webdriver.Chrome('/Users/hyewon_oh/study/chromedriver')
driver.get('https://search.shopping.naver.com/search/category?catId=50000167')
rank=1

for i in range(20):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)

   

    source = driver.page_source
    s1 = BeautifulSoup(source, "html.parser")
    
    list=s1.find_all('li',class_="basicList_item__2XT81")
    
    for item in list:
        title=item.find('div',class_="basicList_title__3P9Q7")
        price=item.find('span',class_="price_num__2WUXn")
        category=item.find('div',class_="basicList_depth__2QIie")
        bottom=item.find_all('span',class_="basicList_etc__2uAYO")
        review=item.find('em',class_="basicList_num__1yXM9")
        mall=item.find('div',class_="basicList_mall_title__3MWFY").find_all('a',class_="basicList_mall__sbVax")


        print("%s위"%rank)
        rank+=1
        print(title.text.strip())
        print(mall[0].text.strip())

        if item.find('span',class_="basicList_mall_name__1XaKA"):
            seller=item.find_all('span',class_="basicList_mall_name__1XaKA")
            print("판매처: ")
            for shop in range(len(seller)):
                if shop == 0: continue
                print(seller[shop].text.strip())


        print(price.text.strip())
        print(category.text.strip())
        print(bottom[0].text.strip())
        print(bottom[1].text.strip())
        print("리뷰 : "+review.text.strip())

        if item.find('div',class_="basicList_detail_box__3ta3h").find('a',class_="basicList_detail__27Krk"):
            detail=item.find('div',class_="basicList_detail_box__3ta3h")
            print(detail.text.strip())


        print("-----------------------------------")
        
    if i == 19 :
        driver.close()
    else :
        driver.find_element_by_class_name('pagination_next__1ITTf').click()
```
