```python
from bs4 import BeautifulSoup
from selenium import webdriver

driver= webdriver.Chrome('/Users/hyewon_oh/study/chromedriver')
driver.get('https://www.coupang.com/np/categories/186764')
n=2
rank=1;

for num in range(1,20):
    url="https://www.coupang.com/np/categories/186764?page=%s" %num
    source = driver.page_source

    s1 = BeautifulSoup(source, "html.parser")

    item = s1.find_all('a', class_='baby-product-link')
    
    for i in item:
        title=i.find('div', class_='name')
        price=i.find('strong', class_='price-value')
        rating=i.find('em', class_='rating')
        reward= i.find('span',class_="reward-cash-txt")
        derivery=i.find('span',class_="arrival-info emphasis")
        derivery=derivery.find_all('em')


        print("%d위"%rank)
        print("상품:"+title.text.strip())
        print("가격:"+price.text.strip())
        print("평점:"+rating.text.strip())
        print("적립 포인트:"+reward.text.strip())
        print("배송 날짜:"+derivery[0].text.strip())

        if i.find('span',class_="badge delivery-info"):
            print("무료 배송")


        if i.find('span', class_='badge rocket'):
            print("로켓배송/제휴 가능")
        else:
            print("로켓배송 불가능")
        
        if i.find('div',class_="out-of-stock"):
            print("품절")
        rank+=1
        print("----------------------------------")

    if num == 19 :
        driver.close()
    
```
