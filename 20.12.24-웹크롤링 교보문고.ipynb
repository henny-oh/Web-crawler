```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver= webdriver.Chrome('/Users/hyewon_oh/study/chromedriver')
driver.get('http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf?orderClick=d79')

rank = 0

for i in range(10):
    

    source = driver.page_source

    s1 = BeautifulSoup(source, "html.parser")

    title = s1.find_all('div', class_='title')[4:]

    title_list=[]
    title = s1.find_all('div', class_='title')

    for item in title[4:]:
        title_list.append(item.text.replace('\n',''))
    title_list

    author_list=[]
    author = s1.find_all('div', class_='author')
    for item in author:
        author_list.append(item.text.replace('\n','').replace('\t',''))
    author_list

    price_list = []
    price = s1.find_all('div', class_='price')
    for item in price:
        price_list.append(item.text.replace('\n','').replace('\t',''))
    price_list

    
    for idx in range(len(title_list)):
        rank +=1
        print('%s위 도서정보' %rank)
        print(title_list[idx])
        print(author_list[idx])
        print(price_list[idx])
        print('-'*80)
    
    if i == 9 :
        driver.close()
    elif i ==0:
        driver.find_element_by_xpath('//*[@id="main_contents"]/div[4]/div[1]/a').click()
    else:
        driver.find_element_by_xpath('//*[@id="main_contents"]/div[4]/div[1]/a[2]').click()
```
