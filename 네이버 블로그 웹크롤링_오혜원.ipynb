```python
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
import time
import warnings
import math
warnings.filterwarnings(action='ignore')

keyword=input('키워드를 입력하세요')
cnt=int(input('크롤링할 개수를 입력하세요'))
pnt=math.ceil(cnt/7)

driver= webdriver.Chrome('/Users/hyewon_oh/study/chromedriver')

pnum=0
num=0
for page in range(1,pnt+1):
    driver.get('https://section.blog.naver.com/Search/Post.nhn?pageNo=%s&rangeType=ALL&orderBy=sim&keyword=%s' %(page,keyword))
    time.sleep(3)
    
    source = driver.page_source
    s1 = BeautifulSoup(source, "html.parser")
    
    posts=s1.find_all('div',class_="list_search_post")
    pnum=0
    for post in posts:
        pnum+=1
        blog_name=post.find('span',class_="name_blog")
        author=post.find('em',class_="name_author")
        title=post.find('span',class_="title")
        date=post.find('span',class_="date")
        driver.find_element_by_xpath('''//*[@id="content"]/section/div[2]/div[%s]/div/div[1]/div/a[2]'''%pnum).click()
        
        driver.switch_to_window(driver.window_handles[-1])
        
        source = driver.page_source
        s1 = BeautifulSoup(source,'html.parser')
        
        blog= s1.find('iframe').get('src')
        response = requests.get('https://blog.naver.com/' + blog)
        source = response.text
        s2 = BeautifulSoup(source,'html.parser')
        
        try:
            content = s2.find('div', class_='se-main-container').text.replace('\n','').replace('\u200b','').replace('  ','').strip()
        except:
            content =  s2.select('#postViewArea')[0].text.replace('\n','').replace('\xa0','').replace('\ufeff','').replace('  ','').strip()
        driver.close()
        driver.switch_to_window(driver.window_handles[0])
        
        num+=1
        print("%s번째 포스트"%num)
        print("author="+author.text.strip())
        print("title="+title.text.strip())
        print("blog_name="+blog_name.text.strip())
        print("date="+ date.text.strip())
        print(content)
        
        if num==cnt:
            break
        
        
        print("---------------------------------")
        
driver.close()
```
