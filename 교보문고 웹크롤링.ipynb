```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver= webdriver.Chrome('/Users/hyewon_oh/study/chromedriver')
driver.get('http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf?orderClick=d79')

rank = 0


for i in range(10):
    source = driver.page_source

    s1 = BeautifulSoup(source, "html.parser")

    item = s1.find_all('div', class_='detail')

    for book in item:

        title=book.find( 'div', class_='title').text.replace('\n','').replace('\t','')
        author=book.find('div',class_='author').text.replace('\n','').replace('\t','')
        price =book.find('div', class_='price').text.replace('\n','').replace('\t','')
       

        rank+=1

        print("%s위"%rank)
        print("제목:"+title)
        print("저자|출판사|출판일: "+author)
        
        if book.find('strong', class_='blue'):
            info=book.find( 'strong', class_='blue').text.replace('\n','').replace('\t','')
            print(info)
        print("---------------------------------")
    if i == 9 :
        driver.close()
    elif i ==0:
        driver.find_element_by_xpath('//*[@id="main_contents"]/div[4]/div[1]/a').click()
    else:
        driver.find_element_by_xpath('//*[@id="main_contents"]/div[4]/div[1]/a[2]').click()
```
